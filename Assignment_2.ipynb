```python
#Problem 1

nums = [4, 6, 9, 3, 1]

if nums[2] > nums[0]:
    print("Valid")
else:
    print("Invalid")
```

```python
#Problem 2

students = ("Alice", "Bob", "Charlie")
marks = {"Alice": 95, "Bob": 78, "Charlie": 60}

score = marks[students[1]]

if score >= 90:
    print("Grade A")
elif score >= 70:
    print("Grade B")
else:
    print("Grade F")
```

```python
#Problem 3

a = {"CS101", "CS102", "CS103"}
b = {"CS103", "CS104"}

common = a & b

if len(common) == 1:
    print("One common course")
else:
    print("No match or multiple matches")
```
