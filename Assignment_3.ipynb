```python
#1)Customer Segmentation Based on Purchase Amount




customers = [
    {"name": "Amit", "purchase": 12000},
    {"name": "Radhika", "purchase": 55000},
    {"name": "John", "purchase": 3500},
    {"name": "Meera", "purchase": 28000},
    {"name": "Karan", "purchase": 8000}
]

for cust in customers:
    amt = cust["purchase"]
    if amt > 50000:
        seg = "Platinum"
    elif 20000 <= amt <= 50000:
        seg = "Gold"
    elif 5000 <= amt < 20000:
        seg = "Silver"
    else:
        seg = "Bronze"
    print(f"{cust['name']}: {seg}")
```

```python
#2)Detect Outliers in a Dataset (Manual IQR Logic)





delivery_times = [22, 25, 30, 28, 27, 120, 26, 24, 23, 200, 29, 28, 26]

def median(sorted_list):
    n = len(sorted_list)
    mid = n // 2
    if n % 2 == 1:
        return sorted_list[mid]
    else:
        return (sorted_list[mid - 1] + sorted_list[mid]) / 2


s = sorted(delivery_times)

n = len(s)
if n % 2 == 0:
    lower_half = s[:n//2]
    upper_half = s[n//2:]
else:
    lower_half = s[:n//2]
    upper_half = s[n//2 + 1:]

Q1 = median(lower_half)
Q3 = median(upper_half)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

print(f"Q1 = {Q1}, Q3 = {Q3}, IQR = {IQR}")
print(f"Outlier bounds: ({lower_bound}, {upper_bound})")

outliers = [x for x in delivery_times if x < lower_bound or x > upper_bound]
print("Outliers:", outliers)
```

```python
#3) Classify Employees Based on Attendance %





attendance = {
    "Suresh": 96,
    "Pooja": 88,
    "Rahul": 72,
    "Sneha": 65,
    "Vikram": 94,
    "Manoj": 81
}

for name, pct in attendance.items():
    if pct >= 95:
        rating = "Excellent"
    elif 85 <= pct <= 94:
        rating = "Good"
    elif 70 <= pct <= 84:
        rating = "Average"
    else:
        rating = "Poor"
    print(f"{name}: {rating}")
```

```python
#4)Count Missing Values Without Using Pandas






data = [23, None, 45, "", 32, None, "NA", 0, " ", 55, "na", "N/A", 100]

missing_count = 0
missing_items = []

for item in data:
    if item is None:
        missing_count += 1
        missing_items.append(item)
    elif isinstance(item, str):
        cleaned = item.strip().lower()
        if cleaned == "" or cleaned in ("na", "n/a"):
            missing_count += 1
            missing_items.append(item)

print("Missing values count:", missing_count)
print("Missing items found:", missing_items)
```

```python
#5) Find Longest Increasing Sequence (Analytical Thinking)





sales = [100, 120, 130, 90, 95, 110, 115, 117]

max_streak = 0
current_streak = 0

for i in range(1, len(sales)):
    if sales[i] > sales[i-1]:

        if current_streak == 0:
            current_streak = 2
        else:
            current_streak += 1
    else:

        if current_streak > max_streak:
            max_streak = current_streak
        current_streak = 0


if current_streak > max_streak:
    max_streak = current_streak


if max_streak == 0 and len(sales) > 0:
    result = 1
else:
    result = max_streak

print("Longest increasing streak (in days):", result)
```

```python
#1) Function to Categorize Age Groups



def age_category(age):
    if age < 13:
        return "Child"
    elif 13 <= age <= 19:
        return "Teen"
    elif 20 <= age <= 59:
        return "Adult"
    else:
        return "Senior"

age_list = [5, 14, 22, 67, 45, 13, 80, 19]
for a in age_list:
    print(a, "->", age_category(a))
```

```python
#2)Function to Filter Products Above Given Rating




product_reviews = [
    {"name": "Earbuds", "rating": 4.3},
    {"name": "Laptop Bag", "rating": 3.8},
    {"name": "Keyboard", "rating": 4.7},
    {"name": "Monitor", "rating": 4.0},
    {"name": "USB Cable", "rating": 2.9}
]

def filter_by_rating(products, threshold):
    result = []
    for p in products:
        if p.get("rating", 0) >= threshold:
            result.append(p)
    return result

good_products = filter_by_rating(product_reviews, 4.0)
print("Products with rating >= 4.0:")
for p in good_products:
    print("-", p["name"], "rating:", p["rating"])
```

```python
#3)Function to Return Length of a String




def string_length(text):
    count = 0
    for _ in text:
        count += 1
    return count

text = "Data Analyst"
print(f"Text: '{text}' has {string_length(text)} characters (manual count).")
```

```python
#5)Function to Square Each Number in a List



def square_list(values):
    squared = []
    for v in values:
        squared.append(v * v)
    return squared

values = [1, 2, 3, 4, 5]
print("Squared:", square_list(values))
```
